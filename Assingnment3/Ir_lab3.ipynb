```python
import string
description={}
def splitting(x):
  table=str.maketrans('','',string.punctuation)
  i=[]
  i=x.split()
  i=[word.lower() for word in i]
  i=[w.translate(table) for w in i]
  i=[w for w in i if w!='']
  return i
```
